```python
import matplotlib.lines as mlines
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Databases/Params.csv', delimiter=',')

df
```

```python
df_struct = df[['Date', 'Personal consumption spending','Investment spending','Government spending','Clear export']]
```

```python
# Decide Colors 
mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive']      

# Draw Plot and Annotate
fig, ax = plt.subplots(1,1,figsize=(16, 9), dpi= 80)
columns = df_struct.columns[1:]
labs = columns.values.tolist()

# Prepare data
x  = df_struct['Date'].values.tolist()
y0 = df_struct[columns[0]].values.tolist()
y1 = df_struct[columns[1]].values.tolist()
y2 = df_struct[columns[2]].values.tolist()
y3 = df_struct[columns[3]].values.tolist()
#y4 = df_struct[columns[4]].values.tolist()

y = np.vstack([y0, y2, y1, y3])

# Plot for each column
labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x, y, labels=labs, colors=mycolors, alpha=0.8)

# Decorations
title = "Structure of aggregate demand in the USA"
ax.set_title('Structure of aggregate demand in the USA', fontsize=18)
ax.set(ylim=[0, 30000])
ax.legend(fontsize=10, ncol=4)
plt.xticks(x[::5], fontsize=10, horizontalalignment='center')
plt.yticks(np.arange(10000, 31000, 10000), fontsize=10)
plt.xlim(x[0], x[-1])

# Lighten borders
plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.3)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.3)

plt.savefig(f'Plots/{title}.png')
plt.show()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
```

```python
counts = pd.read_csv('Databases/VVP.csv', index_col='Date', parse_dates=True)
counts2 = pd.read_csv('Databases/Params.csv', index_col='Date', parse_dates=True)
params = pd.read_csv('Databases/pcs.csv', index_col='Date', parse_dates=True)
```

```python
yearly = counts.resample('d').sum()
yearly['VVP'] = yearly.sum(axis=1)
yearly = yearly[['VVP']]
```

```python
params['pcs'] /= 1
yearly = yearly.join(params[['pcs']])
yearly = yearly.join(counts2[['Investment spending']])
yearly = yearly.join(counts2[['Government spending']])
yearly = yearly.join(counts2[['Clear export']])
```

```python
# Drop any rows with null values
yearly.dropna(axis=0, how='any', inplace=True)

column_names = ['pcs', 'Investment spending', 'Government spending', 'Clear export']
X = yearly[column_names]
y = yearly['VVP']

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
yearly['predicted'] = model.predict(X)
```

```python
yearly[['VVP', 'predicted']].plot(alpha=0.5);
```

```python
params = pd.Series(model.coef_, index=X.columns)
params
```

```python
from sklearn.utils import resample
np.random.seed(1)
err = np.std([model.fit(*resample(X, y)).coef_
              for i in range(1000)], 0)
```

```python
print(pd.DataFrame({'effect': params.round(0),
                    'error': err.round(0)}))
```

```python
infl = pd.read_csv('Databases/infl.csv', index_col='Date', parse_dates=True)
counts = pd.read_csv('Databases/Params.csv', index_col='Date', parse_dates=True)
params = pd.read_csv('Databases/pcs.csv', index_col='Date', parse_dates=True)
```

```python
yearly_cl_ex = counts.resample('d').sum()
yearly_cl_ex = yearly_cl_ex[['Clear export']]
```

```python
yearly_cl_ex = yearly_cl_ex.join(params[['pcs']])
yearly_cl_ex = yearly_cl_ex.join(infl[['Infl']])
yearly_cl_ex = yearly_cl_ex.join(counts[['Investment spending']])
```

```python
# Drop any rows with null values
yearly_cl_ex.dropna(axis=0, how='any', inplace=True)

column_names_cl_ex = [ 'Infl', 'Investment spending', 'pcs']
X = yearly_cl_ex[column_names_cl_ex]
y = yearly_cl_ex['Clear export']

model = LinearRegression(fit_intercept=False)
model.fit(X, y)
yearly_cl_ex['predicted'] = model.predict(X)
```

```python
yearly_cl_ex[['Clear export', 'predicted']].plot(alpha=0.5);
```

```python
params = pd.Series(model.coef_, index=X.columns)
params
```

```python
from sklearn.utils import resample
np.random.seed(1)
err = np.std([model.fit(*resample(X, y)).coef_
              for i in range(1000)], 0)
```

```python
print(pd.DataFrame({'effect': params.round(0),
                    'error': err.round(0)}))
```
