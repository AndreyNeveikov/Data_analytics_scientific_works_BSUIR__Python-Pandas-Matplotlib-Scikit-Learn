```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Databases/Functional_classification.csv', delimiter=',')

df.head()
```

```python
df = pd.read_csv('Databases/Functional_classification.csv', delimiter=',')

mycolors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange', 'tab:brown', 'tab:grey', 'tab:pink', 'tab:olive',
            'tab:purple', 'tab:cyan', '#31D115', '#66CCFF', '#ACE600'] 

#vals
vals = df["Процент"]
labels = df["Функциональная классификация"]
fig, ax = plt.subplots()
ax.pie(vals, pctdistance=1.6, radius=2, colors=mycolors)
ax.axis("equal")

#decors
title = "Структура государственных расходов"
ax = plt.gca()
ax.set_title(f"{title}", fontdict={'size':15})

labels = [f"{n} ({v})" for n,v in zip(labels, vals)]
plt.legend(
bbox_to_anchor = (-0.30, 0.90, 0.30, 0.00),
loc = 'best', labels = labels)
plt.savefig(f'Plots/{title}.png')
plt.show()
```
